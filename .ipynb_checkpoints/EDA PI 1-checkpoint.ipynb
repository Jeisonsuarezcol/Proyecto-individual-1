```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from datetime import datetime
from wordcloud import WordCloud
import missingno as msno
import numpy as np
import pandas as pd
```

```python
df_final = pd.read_csv('datasets/final_movies_dataset.csv')
```

```python
df_final['release_date'] = pd.to_datetime(df_final['release_date'], format='%Y-%m-%d')
```

```python
df_final.head()
```

```python
'''Al importar el dataframe visualizamos sus columnas utilizando la funcin matrix de la librera "missingno" 
para generar una matriz de colores donde los espacios en blanco representan los valores faltantes. 
Esto nos  proporciona una pista sobre variables irrelevantes'''

msno.matrix(df_final)
```

```python
'''Para empezar este Anlisis exploratorio de los datos (EDA) vamos a eliminar las variables que considero son
irrelevantes en este caso.'''

df_final = df_final.drop(columns=['id', 'original_language','release_year','runtime', 'status','tagline'])
```

```python
df_final
```

```python
#Se observa que del dataframe obtenido la nica columna que tiene valores nulos es 'overview' lo cual tiene sentido 
#porque no todas pelculas tendrn una descripcin general.

df_final.isnull().any()
```

```python
round(df_final.describe())
```

```python
#Debido a los avances tecnolgicos y el crecimiento de la poblacin, un anlisis interesante sera ver cuantas pelculas
#se han hecho a travs de las dcadas.


# Agrupamos los datos por dcada y se suman los valores correspondientes.
decadas = df_final['release_date'].dt.year // 10 * 10

agrupado = decadas.value_counts()

agrupado = agrupado.sort_values(ascending=True)


#Se crea la grfica de barras
agrupado.plot(kind='bar',color= 'orange')


plt.xlabel('Dcada')
plt.ylabel('Nmero de pelculas')
plt.title('Pelculas por Dcada')

plt.xlim(left=4)


plt.show()
```

```python
#En el siguiente scatter plot se observa cierta relacin positiva entre la cantidad de valoraciones y la popularidad de la pelcula

plt.scatter(df_final['vote_count'], df_final['popularity'], alpha=0.5)

# Personalizar el scatter plot
plt.title("Relacin entre Popularidad y nmero de valoraciones")
plt.xlabel("nmero de valoraciones")
plt.ylabel("Popularidad")

# Ajustes en el rango del eje x y el eje y
plt.xlim(0, 10000) 
plt.ylim(0, 60)

plt.show()
```

```python
# En el siguiete histograma podemos ver que la calificacin ms frecuente en este dataset se encuentra entre 5 y 7 

revenue_values = df_final['vote_average']

plt.hist(revenue_values, bins=10, edgecolor='black')

plt.title('Distribucin del Puntaje promedio')
plt.xlabel('')
plt.ylabel('Frecuencia')

plt.show()
```

```python
#Para este scatter plot se observa cierta correlacin entre la inversin y la recaudacin esto se hace ms evidente 
#entre mayor sea la invesin.

plt.scatter(df_final['budget'], df_final['revenue'], color='green', alpha=0.5)

# Crear un formateador para los ejes
formatter = ticker.FuncFormatter(lambda x, pos: f'{x/1e6:.0f}M')

# Aplicamos el formateador a los ejes x e y
plt.gca().xaxis.set_major_formatter(formatter)
plt.gca().yaxis.set_major_formatter(formatter)

# Etiquetas y ttulo
plt.xlabel('Inversin')
plt.ylabel('recaudacin')
plt.title('Relacin entre inversin y recaudacin')


plt.xlim(0, 280000000) 
plt.ylim(0, 1750000000)

# Mostrar el grfico
plt.show()
```

```python
#En este grfico de dispersin tambin observa una correlacin entre la inversin y el nmero de valoraciones.

plt.scatter(df_final['vote_count'], df_final['revenue'], color='red', alpha=0.3)

# Crear un formateador para los ejes
formatter = ticker.FuncFormatter(lambda x, pos: f'{x/1e6:.0f}M')

# Aplicamos el formateador a los ejes x e y

plt.gca().yaxis.set_major_formatter(formatter)

# Etiquetas y ttulo
plt.xlabel('Nmero de valoraciones')
plt.ylabel('Inversin')
plt.title('Relacin entre Nmero de valoraciones e Inversin')


plt.xlim(0, 8000) 
plt.ylim(0, 1500000000)

# Mostrar el grfico
plt.show()
```

```python
#El cdigo a continuacin genera una nube de palabras para ver cules son las ms frecuentes en los ttulos y descripciones.

text = ' '.join(df_final['title'].apply(lambda x: x.lower()).values)  

# Genera la nube de palabras
wordcloud = WordCloud(width=800, height=400, max_words=200, background_color='white').generate(text)

# Visualizar 
plt.figure(figsize=(12, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# Paso 3: Preparar los datos para generar la nube de palabras
text = ' '.join(df_final['overview'].dropna().apply(lambda x: x.lower()).values)  

# Genera la nube de palabras
wordcloud = WordCloud(width=800, height=400, max_words=100, background_color='white').generate(text)

# Visualizar 
plt.figure(figsize=(12, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
